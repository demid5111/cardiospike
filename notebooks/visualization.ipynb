```python
import pandas as pd

from cardiospike import TRAIN_DATA_PATH
from cardiospike.utils.visualization import plot_rr
```

```python
train = pd.read_csv(TRAIN_DATA_PATH)
```

```python
train["anomaly_proba"] = train["y"]
```

```python
train.groupby("id").apply()
```

```python
train[(train["x"] < 800) & (train["x"] > 400)]["id"].unique()
```

```python
train["anomaly_proba"] = 0
```

```python
def plot_sample(sample_id):
    this_sample = train[train["id"] == sample_id].copy().reset_index()
    return plot_rr(this_sample, study_id=sample_id)[0]
```

```python
plot_sample(258).update_layout(title_text="Один спайк")
```
